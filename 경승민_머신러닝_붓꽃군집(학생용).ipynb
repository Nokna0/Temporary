```python
# 데이터셋 불러오기
from sklearn.datasets import load_iris
iris = load_iris()    # 붓꽃 데이터 불러오기
iris      # iris 확인해 보기
```

```python
print(iris.feature_names) # 데이터셋 속성명 출력
```

```python
# 데이터셋 살펴보기
import pandas as pd
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)   # 데이터프레임으로 생성하기

#  df 살펴보기
df.info()
```

```python
df.head()     # 상위 5개 살펴보기
```

```python
# 상관관계 분석하기,  현재 df에는 4개의 속성만 있고 정답 label은 없음.
df['target'] = iris.target     # 정답 레이블 추가해보기

corr = df.corr()      # 상관관계 구하기
corr      # 상관관계 출력
```

```python
# 속성 추출하기
import matplotlib.pyplot as plt

iris_petal = iris.data[:, [2, 3]]       # 꽃잎 길이와 꽃잎 너비를 iris_petal에 저장
labels = iris.target                # 품종(target)을 labels에 저장
print(iris_petal, labels)          #  iris_petal, labels 출력해 보기
```

```python
# 데이터 시각화하기
import matplotlib.pyplot as plt
plt.scatter(iris_petal[:, 0], iris_petal[:, 1], c ='b', alpha = 0.5)
plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```

```python
# 모델 학습하기
from sklearn.cluster import KMeans
KMean = KMeans(n_clusters = 3)       # k-means 모델 생성
KMean.fit(iris_petal)              # k-means 모델 학습
```

```python
# 결과 확인하기
print(KMean.cluster_centers_)     # 중심값 좌푯값
print(KMean.labels_)              # 군집 구분(군집 결과)
```

```python
# 결과 시각화하기
colours = ['tab:red','tab:blue', 'tab:green']
# tab:orange 등 tab이 붙는 10개의 색상 세트는 시각적으로 구분이 잘되고, 채도 적당, 여러 색을 동시에 써도 가독성이 좋음.
center_colours = ['r', 'b', 'g']

for i in range(len(KMean.labels_)):
  label = KMean.labels_[i]
  plt.scatter(iris_petal[i][0],
              iris_petal[i][1],
              c = colours[label],            # label에 저장해 둔 색상으로 지정하기
              alpha = 0.5)

for i in range(len(KMean.cluster_centers_)):
  plt.scatter(KMean.cluster_centers_[i][0],
              KMean.cluster_centers_[i][1],
              s = 200,           # 표시 사이즈 지정
              c = center_colours[i])   # 표시 색상 지정

plt.xlabel('petal length')
plt.ylabel('petal width')
plt.show()
```

```python
# 최적의 k값과 비교하기

inertia_a = []
for k in range(1, 10):
    kmeans= KMeans(n_clusters=k)
    kmeans.fit(iris_petal)
    inertia = kmeans.inertia_ # 이너셔 값 변수에 저장, 이너셔란 군집의 중심과 군집에 속한 데이터 사이의 거래를 제곱 합하여 계산
    inertia_a.append(inertia) # 이너셔 값(변수) 리스트 inertia_a에 추가

plt.plot(range(1, 10), inertia_a, marker='o')   # 선그래프로 표현
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.show()
```
